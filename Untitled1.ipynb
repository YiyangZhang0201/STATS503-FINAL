```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
value_ls = np.load("valuels.npy")
value_ls = value_ls.tolist()
value_rf = np.load("valuerf.npy")
value_rf = value_rf.tolist()
value_xg = np.load("valuexg.npy")
value_xg = value_xg.tolist()
```

```python
import matplotlib.dates as mdates
Index = pd.read_excel("data/index.xlsx")
Index.columns = ["Date", "prior_close", "Open", "High", "Low", "Close", "volume", 
                 "amount", "change", "change_percent"]
Index = Index[["Date",'Open', 'High', 'Low', 'Close']]
Index["Date"] = mdates.date2num(Index["Date"].values)
plt.figure(figsize=(15,5))
plt.plot(Index["Close"].iloc[1497:1742])
plt.show()
```

```python
Index
```

```python
coe = 1000000/Index["Open"].iloc[1497]
Index["Close"] = Index["Close"]*coe
Index_list = Index["Close"].iloc[1497:1742].values.tolist()
Index_value = [1000000]
Index_value.extend(Index_list)
```

```python
print(Index_list)
```

```python
from mpl_finance import candlestick_ohlc
 
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15,3), dpi=80)

# 第一个参数axes
candlestick_ohlc(axes, Index.values, width=0.3, colorup="r", colordown="g")
axes.xaxis_date()
plt.title("SSE 50 Index")
plt.xlabel("Time")
plt.ylabel("Index Value")
plt.savefig("sse50index.png")
```

```python
plt.figure(figsize=(15,4))
plt.plot(Index_value, label="Index")
plt.plot(value_ls, label="Lasso Portfolio")
plt.plot(value_xg, label="XGBoost Portfolio")
plt.plot(value_rf, label="Random Forest Portfolio")
plt.legend()
plt.xlabel("Number of Traiding Days")
plt.ylabel("Portfolio Value")
plt.title("Index Value v.s. Factor Based Portfolio")
plt.savefig("backtestresultfinal")
```

```python
Index0 = pd.read_csv("cleaned_stock/600000.csv")
Index1 = pd.read_csv("cleaned_stock/600519.csv")
Index0 = Index0[["time",'open', 'high', 'low', 'close']]
Index1 = Index1[["time",'open', 'high', 'low', 'close']]
Index0["time"] = mdates.date2num(Index0["time"].values)
Index1["time"] = mdates.date2num(Index1["time"].values)

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15,4), dpi=80)

# 第一个参数axes
candlestick_ohlc(axes[0], Index0.values, width=0.3, colorup="r", colordown="g")
candlestick_ohlc(axes[1], Index1.values, width=0.3, colorup="r", colordown="g")
axes[0].xaxis_date()
axes[1].xaxis_date()
fig.suptitle("600000.SH & 600519.SH Stock Price Plot")
axes[0].set_title('600000.SH', y=1.0, pad=-14)
axes[1].set_title('600519.SH', y=1.0, pad=-14)
# axes.xlabel("Time")
axes[0].set_ylabel("Stock Price")
axes[1].set_ylabel("Stock Price")

plt.savefig("sp.png")
```

```python
import seaborn as sns
import statistics as stat
data = pd.read_csv("data_with_factor/600000.csv")
sns.histplot(data["Target"], bins=40, kde=True)
plt.xlabel("Target")
plt.axvline(x=stat.mean(data["Target"]),color='red')
plt.title("600000.SH Target Distribution")
plt.savefig("targetdistribution00")
```
